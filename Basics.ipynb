#### explore_data() function takes a dataset in the form of a list of lists, and prints out a selection of that dataset given starting and ending points as parameters. Optionally, it can print out the row and column numbers.

```python
def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]
    for row in dataset_slice:
        print(row)
        print('\n')

    if rows_and_columns:
        print('Number of rows: ', len(dataset))
        print('Number of column: ', len(dataset[0]))
```

```python
# Create two lists: 1 for unique and 1 for duplicate apps; the list for duplicate apps
# should not have duplicates itself; otherwise, care would have to be taken to make
# sure that 1 row in the dataset with a name in the duplicates list is kept; IE: we do
# not want to remove all the rows that are duplicates; we want to keep 1
g_dup_apps = []
g_unique_apps = []

for row in google_data:
    name = row[0]
    # This if clause keeps duplicates from occurring in the duplicates list
    if name in g_unique_apps and name in g_dup_apps: continue
    elif name not in g_unique_apps: g_unique_apps.append(name)
    elif name in g_unique_apps and name not in g_dup_apps: g_dup_apps.append(name)

print('Number of google apps that have duplicates: ', len(g_dup_apps))
print('\n')
print('Print first 10 duplicates:\n', g_dup_apps[:10])
```

#### The apps with the highest number of downloads are kept in google_data set. All others are deleted.

```python
# remove_dups()
# Function takes in the name of an app w/ duplicates, a dataset, a list, and a
# column number as parameters. It creates a list of the row numbers to be deleted.
# These rows are sorted from least to greatest values of the column number that is
# passed as an argument. The row with the greatest of those values is not deleted. 
def remove_dups(an_app, a_dataset, a_col):

    # Iterate through a_dataset; if the app name matches the one passed as an argument,
    # append its index number to the row, and append the row to dup_list
    dup_list = []
    for i in range(0, len(a_dataset)):
        row = a_dataset[i]
        if row[0] == an_app:
            row.append(i)
            dup_list.append(row)

    # Sort dup_list from greatest to least value in position a_col; delete the row with
    # the greatest value from dup_list
    dup_list.sort(key=lambda x: x[a_col], reverse=True)
    del dup_list[0]

    # Iterate through dup_list; delete each row in a_dataset that corresponds to the index
    # appended to each list in dup_list
    j = 0
    while j < len(dup_list):
        del a_dataset[j]
        j += 1

    return(a_dataset)
```

```python
# remove_non_eng()
# This function take in a string and returns False if there are non-english characters
# in it; otherwise, it returns True; if the ASCII is greater than 127, then it is
# non-english; no return means that all characters in the string are english
def bool_non_eng(a_string):
    count = 0
    for char in a_string:
        if ord(char) > 127: count += 1
    if count > 3: return False
    else: return True
```

#### Genre Frequency Tables

```python
# freq_table()
# Function takes a dataset, an index to count data in, and it has a second index parameter set as
# by default to None; if b_index == None, it returns a list of data pertaining to the values in
# the a_index; this is a dictionary of frequencies followed by figure of the total count to be
# used in percentage calculations; If b_index != None, it returns a list containing a_table,
# a_total, b_table, b_total, ab_table, and ab_total
def freq_table(a_dataset, a_index, b_index=None):
    
    a_table = {}
    b_table = {}
    ab_table = {}

    # Count the occurrences
    for row in a_dataset:
        if row[a_index] in a_table: a_table[row[a_index]] += 1
        else: a_table[row[a_index]] = 1
        if b_index != None:    
            if row[b_index] in b_table: b_table[row[b_index]] += 1
            else: b_table[row[b_index]] = 1
            ab_combined = row[a_index] + ", " + row[b_index]
            if ab_combined in ab_table: ab_table[ab_combined] += 1
            else: ab_table[ab_combined] = 1
    
    # To store the sume of all occurrences
    a_sum = 0
    b_sum = 0
    ab_sum = 0
    
    # Sum values in a_table
    for key in a_table:
        a_sum += a_table[key]
    
    # If argument given for b_index, sum b and ab tables
    if b_index != None:
        for key in b_table:
            b_sum += b_table[key]
        for key in ab_table:
            ab_sum += ab_table[key]
    
    # Change values in a_table to frequencies
    for key in a_table:
        a_table[key] = (a_table[key] / a_sum) * 100
    
    # If value passed for b_index, change values in b and ab tables to freq's
    if b_index != None:
        for key in b_table:
            b_table[key] = (b_table[key] / b_sum) * 100
        for key in ab_table:
            ab_table[key] = (ab_table[key] / ab_sum) * 100
            
    # If b_index == None, return list of a data
    if b_index == None:
        a_list = [a_table]
        return a_list
    else:
        # If b_index != None, return all a, b, and ab data
        ab_list = [a_table, b_table, ab_table]
        return ab_list
```

```python
def print_table(a_dataset, a_index, b_index=None):

    # Call freq_table() function w/ values passed; save output
    list_tables = freq_table(a_dataset, a_index, b_index)

    # Separate a_list just in case there are others
    a_list = list_tables[0]

    # Create list of tuples from dictionary
    a_freqs = []
    for key in a_list:
        tup = (a_list[key], key)
        a_freqs.append(tup)

    # Sort a_freqs
    a_freqs = sorted(a_freqs, reverse=True)

    # Print items in sorted a_list
    print("a_index Frequencies:")
    print(20 * "-", "\n")
    for item in a_freqs:
        print(item[1], " : ", item[0])
    print("\n")

    # Check to see if b_index value was passed; sort; print
    if len(list_tables) > 1:

        # Create lists to print freq's
        b_freqs = []
        ab_freqs = []

        # Append dictionary keys, vals to b_list
        b_list = list_tables[1]
        for key in b_list:
            tup = (b_list[key], key)
            b_freqs.append(tup)

        # Sort b_freqs
        b_freqs = sorted(b_freqs, reverse=True)

        # Print items in b_list
        print("b_index Frequencies:")
        print(20 * "-", "\n")
        for item in b_freqs:
            print(item[1], " : ", item[0])
        print("\n")

        # Append dictionary keys, vals to ab_list
        ab_list = list_tables[2]
        for key in ab_list:
            tup = (ab_list[key], key)
            ab_freqs.append(tup)

        # Sort ab_freqs
        ab_freqs = sorted(ab_freqs, reverse=True)

        # Print items in ab_list
        print("ab_index Frequencies:")
        print(21 * "-", "\n")
        for item in ab_freqs:
            print(item[1], " : ", item[0])
```
